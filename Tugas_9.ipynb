```python
# Latihan
# 1. Diberikan list berikut:
# buah = ['apel', 'mangga', 'jeruk', 'anggur', 'pisang']
# a. Tambahkan 'semangka' di akhir list
# b. Sisipkan 'durian' di antara 'jeruk' dan 'anggur'
# c. Hapus 'mangga' dari list
# d. Ubah 'pisang' menjadi 'nanas'
# e. Tampilkan 3 buah pertama

buah = ['apel', 'mangga', 'jeruk', 'anggur', 'pisang']
print(f"list buah : {buah}\n")

buah.append('semangka')
print(f"list buah setelah ditambah semangka : {buah}\n")

buah.insert(3,'durian')
print(f"list buah setelah disisipkan durian : {buah}\n")

buah.remove('mangga')
print(f"list buah setelah mangga dihapus : {buah}\n")

buah[4] = 'nanas'
print(f"mengubah 'pisang' menjadi 'nanas' : {buah}\n")

buah_0 = buah[:3]
print(f"menampilkan 3 buah pertama : {buah_0}")
```

```python
# 2. Diberikan list berikut:
# angka = [45, 12, 78, 23, 56, 89, 34]
# a. Urutkan list secara ascending
# b. Urutkan list secara descending

angka = [45, 12, 78, 23, 56, 89, 34]
print(f"daftar angka : {angka}")

# ascending (dari terkecil ke terbesar)
angka.sort()
print(f"daftar angka ascending : {angka}")

# descending (dari terbesar ke terkecil)
angka.reverse()
print(f"daftar angka descending : {angka}")
```
